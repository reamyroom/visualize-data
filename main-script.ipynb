# Project : Data Visualization California Cities

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
cities = pd.read_csv('./data/california_cities.csv')
cities.head()
cities['area_total_km2'].sort_values(ascending=False)
```

```python
#váº­y chia 100 cho area_range
```

```python
lat, lon = cities['latd'], cities['longd']

population, area = cities['population_total'], cities['area_total_km2']

plt.figure(figsize=(10,8))
plt.scatter(lon, lat, c=np.log10(population), cmap='viridis',
           s=area, linewidth = 1, alpha = 0.75);
plt.axis('equal');
plt.xlabel('longitude');
plt.ylabel('latitude');
plt.colorbar(label='log$_{10}$(population)'); #toshow colorbar on the right
plt.clim(3,7)

plt.title('California Cities : Population and Area Distribution')

#create a legend for cities' size
area_range =[50, 100, 300, 500]
for area in area_range :
    plt.scatter([],[], s=area, label=str(area)+'km$^2$') # let 2 empty list to pass argument label in
plt.legend(labelspacing=2, title = 'City Area')# thÃªm scatter point = 1 nghÄ©a lÃ  vá»i má»i 50km^2 thÃ¬ ta chá» cáº§n
                                                                #1 Äiá»m kÃ­ch thÆ°á»c 1 thÃ´i
```
